```python
# Function to read info from CSV and return the required metrics
def ratio_to_level(percent):
    if percent < 0.2:
        return 0
    elif percent < 0.5:
        return 1
    elif percent < 0.7:
        return 2
    elif percent < 1:
        return 3
    else:
        return 4

def check_differ(level, ratio):
    if level == 0:
        if ratio <= 0.25:
            return 0
        else:
            return abs(level - ratio_to_level(ratio))
    elif level == 1:
        if ratio >= 0.15 and ratio <= 0.55:
            return 0
        else:
            return abs(level - ratio_to_level(ratio))
    elif level == 2:
        if ratio >= 0.45 and ratio <= 0.75:
            return 0
        else:
            return abs(level - ratio_to_level(ratio))
    elif level == 3:
        if ratio >= 0.65:
            return 0
        else:
            return abs(level - ratio_to_level(ratio))
    else:
        if ratio >= 0.95:
            return 0
        else:
            return abs(level - ratio_to_level(ratio))
        
def check_truth(level, ratio, case):
    if case == -3 and level == 0:
        return np.nan
    elif case == -3 and level != 0:
        if ratio >= 0.15:
            return 1
        else:
            return 0
    elif case == -2 and level != 0:
        return np.nan
    else:
        is_correct = 0
        
        if level == 0:
            if ratio <= 0.25:
                is_correct = 1
            else:
                is_correct = 0
        elif level == 1:
            if ratio >= 0.15 and ratio <= 0.55:
                is_correct = 1
            else:
                is_correct = 0
        elif level == 2:
            if ratio >= 0.45 and ratio <= 0.75:
                is_correct = 1
            else:
                is_correct = 0
        elif level == 3:
            if ratio >= 0.65:
                is_correct = 1
            else:
                is_correct = 0
        else:
            if ratio >= 0.95:
                is_correct = 1
            else:
                is_correct = 0
        
        if case in [-3, -1, -2] or level == case:
            return is_correct
        else:
            return np.nan
        
def read_info_csv(index, artery_indices):
    metrics = {}
    for i in artery_indices:
        info_dir = f"""C:/Users/nguc4116/Desktop/artery_reconstruction/info_files/{str(index)}/measure_output_{str(i)}.csv"""
        if os.path.isfile(info_dir):
            df = pd.read_csv(info_dir)
            start_idx = 10
            end_idx = -10
            avg_diameter = df['avg_radius'].mean()
            metrics[f'Min_diameter_{i}'] = df['min_distances'][start_idx:end_idx].min()
            metrics[f'Avg_diameter_{i}'] = df['avg_radius'][start_idx:end_idx].mean()
            
            metrics[f'Stenosis_ratio_min_{i}'] = df['stenosis_ratio_min'][start_idx:end_idx].max()
            metrics[f'Stenosis_ratio_avg_{i}'] = df['stenosis_ratio_avg'][start_idx:end_idx].max()
            metrics[f'Stenosis_ratio_min_avg_all_{i}'] = (1-df['min_distances'][start_idx:end_idx]/avg_diameter).max()
            metrics[f'Stenosis_ratio_avg_avg_all_{i}'] = (1-df['avg_radius'][start_idx:end_idx]/avg_diameter).max()
            
            metrics[f'Stenosis_level_min_{i}'] = ratio_to_level(df['stenosis_ratio_min'][start_idx:end_idx].max())
            metrics[f'Stenosis_level_avg_{i}'] = ratio_to_level(df['stenosis_ratio_avg'][start_idx:end_idx].max())
            metrics[f'Stenosis_level_min_avg_all_{i}'] = ratio_to_level((1-df['min_distances'][start_idx:end_idx]/avg_diameter).max())
            metrics[f'Stenosis_level_avg_avg_all_{i}'] = ratio_to_level((1-df['avg_radius'][start_idx:end_idx]/avg_diameter).max())
            
            
        else:
            metrics[f'Min_diameter_{i}'] = None
            metrics[f'Avg_diameter_{i}'] = None
            metrics[f'Stenosis_ratio_min_{i}'] = None
            metrics[f'Stenosis_ratio_avg_{i}'] = None
            metrics[f'Stenosis_ratio_min_avg_all_{i}'] = None
            metrics[f'Stenosis_ratio_avg_avg_all_{i}'] = None
            metrics[f'Stenosis_level_min_{i}'] = None
            metrics[f'Stenosis_level_avg_{i}'] = None
            metrics[f'Stenosis_level_min_avg_all_{i}'] = None
            metrics[f'Stenosis_level_avg_avg_all_{i}'] = None
    return metrics
```

```python
# # List of indices
dataset_dir = 'C:/Users/nguc4116/Desktop/artery_reconstruction/info_files/'
pattern = re.compile(r'(\d+)')
sub_nums = []
considered_case = 4

# Iterate over the files in the directory
for filename in os.listdir(dataset_dir):
    match = pattern.match(filename)
    if match:
        index = match.group(1)
        sub_nums.append(index)
        print(f'Extracted INDEX: {index} from filename: {filename}')

# Initialize an empty list to collect the results
indices = sub_nums
results = []
artery_indices = [1, 2, 3]

# Loop through each index and collect the metrics
for index in indices:
    print(f'Processing Sub {index}')
    metrics = read_info_csv(index, artery_indices)
    metrics['Index'] = index
    results.append(metrics)

# Convert the results list to a DataFrame
df_results = pd.DataFrame(results)
print(df_results)
```

```python
# Export the DataFrame to a CSV file
output_file = "C:/Users/nguc4116/Desktop/artery_reconstruction/combined_metrics.csv"
df_results.to_csv(output_file, index=False)
print(f"Metrics have been combined and exported to {output_file}")
```

```python
neurologist_df = pd.read_csv('C:/Users/nguc4116/Desktop/artery_reconstruction/stenosis.csv', sep=',')
name_mapping = {
    1: 'lica',
    2: 'rica',
    3: 'ba'
}

compared_df = {}
selected_col_names = ['ID']
for art_index in artery_indices:
    compared_df[art_index] = {}
    diam_col_name = name_mapping[art_index] + '_diam'
    sten_col_name = name_mapping[art_index] + '_stenosis'
    selected_col_names.append(diam_col_name)
    selected_col_names.append(sten_col_name)
    
    compared_df[art_index]['min_diam_err'] = []
    compared_df[art_index]['avg_diam_err'] = []
    compared_df[art_index]['distal_min_sten_level'] = []
    compared_df[art_index]['avg_min_sten_level'] = []
    compared_df[art_index]['distal_avg_sten_level'] = []
    compared_df[art_index]['avg_avg_sten_level'] = []
    compared_df[art_index]['distal_min_sten_success'] = []
    compared_df[art_index]['avg_min_sten_success'] = []
    compared_df[art_index]['distal_avg_sten_success'] = []
    compared_df[art_index]['avg_avg_sten_success'] = []

keep_df = neurologist_df[selected_col_names]
df_results['Index'] = df_results['Index'].astype(int)

# Merge the DataFrames
merged_df = pd.merge(df_results, keep_df, left_on='Index', right_on='ID')
merged_df = merged_df.drop(columns=['ID'])

output_file = "C:/Users/nguc4116/Desktop/artery_reconstruction/merged_metrics.csv"
merged_df.to_csv(output_file, index=False)
print(f"Metrics have been combined and exported to {output_file}")


for index in indices:
    selected_row = neurologist_df[neurologist_df['ID'] == int(index)].iloc[0]
    selected_result_row = df_results[df_results['Index'] == int(index)].iloc[0]
    
    for art_index in artery_indices:
        diam_col_name = name_mapping[art_index] + '_diam'
        sten_col_name = name_mapping[art_index] + '_stenosis'

        diam = float(selected_row[diam_col_name])
        sten = float(selected_row[sten_col_name])
        
        compared_df[art_index]['min_diam_err'].append(abs(diam-selected_result_row[f'Min_diameter_{art_index}']))
        compared_df[art_index]['avg_diam_err'].append(abs(diam-selected_result_row[f'Avg_diameter_{art_index}']))
        compared_df[art_index]['distal_min_sten_level'].append(check_differ(sten, selected_result_row[f'Stenosis_ratio_min_{art_index}']))
        compared_df[art_index]['avg_min_sten_level'].append(check_differ(sten,selected_result_row[f'Stenosis_ratio_min_avg_all_{art_index}']))
        compared_df[art_index]['distal_avg_sten_level'].append(check_differ(sten, selected_result_row[f'Stenosis_ratio_avg_{art_index}']))
        compared_df[art_index]['avg_avg_sten_level'].append(check_differ(sten, selected_result_row[f'Stenosis_ratio_avg_avg_all_{art_index}']))
        compared_df[art_index]['distal_min_sten_success'].append(check_truth(sten, selected_result_row[f'Stenosis_ratio_min_{art_index}'], considered_case))
        compared_df[art_index]['avg_min_sten_success'].append(check_truth(sten,selected_result_row[f'Stenosis_ratio_min_avg_all_{art_index}'], considered_case))
        compared_df[art_index]['distal_avg_sten_success'].append(check_truth(sten, selected_result_row[f'Stenosis_ratio_avg_{art_index}'], considered_case))
        compared_df[art_index]['avg_avg_sten_success'].append(check_truth(sten, selected_result_row[f'Stenosis_ratio_avg_avg_all_{art_index}'], considered_case))
        
        
        
# print(compared_df)
```

```python
num_valid_samples = {}
total_samples = 0
for art_index in artery_indices:
    print('Artery_' + str(art_index))
    for idx, value in enumerate(compared_df[art_index]['avg_min_sten_success']):
        if value == 0:
            print(indices[idx])
            
    num_sample = np.sum(~np.isnan(np.array(compared_df[art_index]['avg_min_sten_success'])))
    total_samples += num_sample
    num_valid_samples[art_index] = num_sample or 0
    
print(num_valid_samples)
```

```python
for art_index in artery_indices:
    compared_df[art_index]['min_diam_err'] = np.nanmean(np.array(compared_df[art_index]['min_diam_err']))
    compared_df[art_index]['avg_diam_err'] = np.nanmean(np.array(compared_df[art_index]['avg_diam_err']))
    compared_df[art_index]['distal_min_sten_level'] = np.nanmean(np.array(compared_df[art_index]['distal_min_sten_level']))
    compared_df[art_index]['avg_min_sten_level'] = np.nanmean(np.array(compared_df[art_index]['avg_min_sten_level']))
    compared_df[art_index]['distal_avg_sten_level'] = np.nanmean(np.array(compared_df[art_index]['distal_avg_sten_level']))
    compared_df[art_index]['avg_avg_sten_level'] = np.nanmean(np.array(compared_df[art_index]['avg_avg_sten_level']))
    compared_df[art_index]['distal_min_sten_success'] = np.nanmean(np.array(compared_df[art_index]['distal_min_sten_success'])) or 0
    compared_df[art_index]['avg_min_sten_success'] = np.nanmean(np.array(compared_df[art_index]['avg_min_sten_success'])) or 0
    compared_df[art_index]['distal_avg_sten_success'] = np.nanmean(np.array(compared_df[art_index]['distal_avg_sten_success'])) or 0
    compared_df[art_index]['avg_avg_sten_success'] = np.nanmean(np.array(compared_df[art_index]['avg_avg_sten_success'])) or 0
    
print(compared_df)
```
