```python
import numpy as np
import json
import trimesh as tm
```

```python
info_dir = 'C:/Users/nguc4116/Desktop/artery_reconstruction/info_files/6477/'
artery_index = 1
smooth_points = np.genfromtxt(info_dir + f'smooth_points_{artery_index}.txt', delimiter=',')
vmtk_boundary_vertices = np.genfromtxt(info_dir + f'vmtk_boundary_vertices_{artery_index}.txt', delimiter=',')
vmtk_boundary_vertices = np.round(vmtk_boundary_vertices, 2)
vmtk_boundary_faces = np.genfromtxt(info_dir + f'vmtk_boundary_faces_{artery_index}.txt', delimiter=',', dtype=int)
vmtk_boundary_vertices, inverse_indices = np.unique(vmtk_boundary_vertices, axis=0, return_inverse=True)
vmtk_boundary_faces = np.array([inverse_indices[face] for face in vmtk_boundary_faces])

with open(info_dir + f'smooth_connected_lines_{artery_index}.json', 'r') as file:
    smooth_connected_lines = json.load(file)
```

```python
mesh = tm.Trimesh(vertices=vmtk_boundary_vertices, faces=vmtk_boundary_faces)
```

```python
# tm.intersections.mesh_multiplane(mesh, plane_origin, plane_normal, heights)
main_line = smooth_connected_lines[0]

for i in range(0, len(main_line)-1, 10):
    original_point = smooth_points[main_line[i]]
    next_point = smooth_points[main_line[i+1]]
    plane_normal = original_point - next_point
    slice2d = mesh.section(
        plane_normal=plane_normal,
        plane_origin=original_point,

    )
    slice_2D, to_3d = slice2d.to_planar()
    # Convert the 3D point to homogeneous coordinates (4x1)
    point_3d_homogeneous = np.append(original_point, 1)

    # Compute the inverse of the transformation matrix
    inverse_matrix = np.linalg.inv(to_3d)

    # Apply the inverse transformation
    point_2d_homogeneous = np.dot(inverse_matrix, point_3d_homogeneous)
    point_2d = point_2d_homogeneous[:2] / point_2d_homogeneous[3]
    segments = []
    lines = slice_2D.entities
    vertices = slice_2D.vertices
    for line in lines:
        points = line.points
        for i in range(0, len(points)):
            point_1 = points[i]
            point_2 = points[i-1]
            segments.append([vertices[point_1].view(np.ndarray) , vertices[point_2].view(np.ndarray)])
    line_segments = (np.array(segments))
    # Create a plot
    plt.figure()

    # Plot the line segments
    for segment in line_segments:
        plt.plot(segment[:, 0], segment[:, 1], 'b-')

    # Plot the 2D point
    plt.scatter(point_2d[0], point_2d[1], color='red', label='2D Point')

    # Add labels and legend
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.legend()
    plt.title('2D Point and Line Segments')

    plt.show()
```
